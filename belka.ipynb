```python
from sympy import *
import sympy as sym
from sympy.physics.mechanics import *
from sympy import init_printing
from scipy.optimize import root_scalar
mechanics_printing(pretty_print=True)
import numpy as np

t=Symbol('t')
C1, C2, C3, C4, k, L0 =symbols('C1 C2 C3 C4 k L0',positive=True)
x=dynamicsymbols('x')
dx=dynamicsymbols('x',1)
d2x=dynamicsymbols('x',2)

X = C1*exp(-k*x)+C2*exp(k*x)+C3*sin(k*x)+C4*cos(k*x)
dX = diff(X, x)
d2X = diff(X, x, 2)
d3X = diff(X, x, 3)


WB1 = X.subs({x:0}).doit()
WB2 = X.subs({x:9*L0})
WB3 = d2X.subs({x:0})
WB4 = d2X.subs({x:9*L0})

SWB1, w = sym.linear_eq_to_matrix(WB1, C1, C2, C3, C4)
SWB2, w = sym.linear_eq_to_matrix(WB2, C1, C2, C3, C4)
SWB3, w = sym.linear_eq_to_matrix(WB3, C1, C2, C3, C4)
SWB4, w = sym.linear_eq_to_matrix(WB4, C1, C2, C3, C4)

COEF = Matrix.vstack(SWB1, SWB2, SWB3, SWB4)

COEF

WYZN = sym.det(COEF)
WYZN.simplify()


```
