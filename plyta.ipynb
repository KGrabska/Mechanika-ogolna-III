```python
from sympy import *
import sympy as sym
from sympy.physics.mechanics import *
from sympy import init_printing
from scipy.optimize import root_scalar
mechanics_printing(pretty_print=True)
import numpy as np

t=Symbol('t')
nu, C1, C2, p0, r0, Dh, r, m0, c, q, ra, rb =symbols('nu C1 C2 p0 r0 D_h r m0 c q r_a r_b',positive=True)
psi = symbols('psi')

psi = -(q*r**3)/(16*Dh)+C1*r+C2/r



E1 = Eq((C1*(1+nu)+C2*((nu-1)/r**2)),m0/Dh).subs({r:0.9*r0})
E2 = Eq((C1*(1+nu)+C2*((nu-1)/r**2)),0).subs({r:r0})

C1r = solve(E1,C1)
C2r = E2.subs(C1,C1r[0])
C2w = solve(C2r, C2)
C1w = C1r[0].subs({C2:C2w[0]})
C2w = C2w[0]

psiw = psi.subs([(C1, C1w),(C2, C2w), (q, 0)])

mr = Dh*(diff(psiw, r)+nu*(psiw/r))
MR = mr.simplify()

mfi = Dh*(psiw/r+nu*diff(psiw,r))
MF = mfi.simplify()

```
